```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
titanic = pd.read_csv("/content/Titanic-Dataset.csv")
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
mean_value = titanic['Age'].mean()
titanic['Age'].fillna(mean_value,inplace = True)
```

```python
titanic.isnull().sum()
```

```python
titanic_clean = titanic[['Pclass','Sex','Age','Fare']]
```

```python
titanic_clean
```

```python
titanic_target = titanic[['Survived']]
titanic_target
```

```python
sex_change = {'male': 1, 'female':2}
titanic_clean['Sex'] = titanic_clean['Sex'].replace(sex_change)
titanic_clean
```

```python
titanic_clean['Age'] = titanic_clean['Age'].astype(int)
titanic_clean
```

```python
x_train,x_test,y_train,y_test = train_test_split(titanic_clean,titanic_target,test_size = 0.2, random_state =60)
```

```python
y_train
```

```python
tree = DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
y_pred = tree.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
tree.score(x_test,y_test)
```

```python
forest = RandomForestClassifier()
forest.fit(x_train,y_train)
y_pred2 = forest.predict(x_test)

y_pred[0]
```

```python
accuracy_score(y_test,y_pred2)
```
